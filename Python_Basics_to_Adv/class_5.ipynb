* Print only the integers without duplictaes and sort them in a descending order

```python
L = [5, 2, 78, "54", 34.7, "98sp", "2", 78]

res=[]
for x in L:
    if type(x)==int and x not in res:
        res.append(x)

'''
x = 5, res = [5]
x = 2, res = [5,2]
x = 78, res = [5,2,78]
x = "54", res = [5,2,78]
x = 34.7, res = [5,2,78]
x = "98sp", res = [5,2,78]
x = "2", res = [5,2,78]
x = 78, res = [5,2,78]
'''

print(res)

res.sort()
print(res)

res.sort(reverse=True)
print(res)
```
