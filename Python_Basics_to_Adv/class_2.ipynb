### String

```python
s = "Sai"

print("s :", s)
print("length of s :", len(s))
```

### accessing string elements

```python
print(s[0])
print(s[1])
print(s[2])
# print(s[3])
```

```python
s = "Sai Preetham reddy udumula"

print(s[-1])
print(s[-2])
print(s[-3])

print(s[3])
```

```python
print(s[len(s) - 1])      # or s[-1]     ---> accessing the last element using 'len' keyword.
```

### String slicing

```python
print(s[0:2])                        # 0:2 ---> 0 is included, 2 is excluded => 0,1
print(s[0:3], len(s[0:3]))
print(s[0:4], len(s[0:4]))

print(s[3:10])
print(s[4:15])
```

```python
print(s[9:3])            # the 1st number should be the smaller and the 2nd number should be the bigger
```

```python
print(s[-1:-3])
```

```python
print(s[-3:-1])
```

```python
s = "sai preetham"

'''
0 -> "s"
1 -> "a"
2 -> "i"

-1 -> "m"
-2 -> "a"
-3 -> "h"
-4 -> "t"
-5 -> "e"
'''

s[2:-5]
```

### string methods

```python
s = "Krishna Chaitanya Reddy Udumula"
```

```python
print(s.upper())       # converts all the characters to capital letters
print(s.lower())       # converts all the characters to small letters
print(s.capitalize())  # only the first letter is capital
print(s.title())       # the first letter of each word is capital
```

```python
print(s.count("s"))
print(s.count("a"))
print(s.count("k"))
```

```python
print(s.index("r"))
print(s.index("C"))
```

```python
s = "Subhashini"
```

```python
print(s.isalpha())
```

```python
print(s.isnumeric())
```

```python
print(s.isnumeric())
```

### split method

```python
s = "Sai Preetham"

print(s.split())        # default value of split is " "
print(s.split("a"))
```

### strip method

```python
s = "        Sai Preetham                        "

print(s, len(s))

s1 = s.strip()
print(s1, len(s1))
```
